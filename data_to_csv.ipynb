```python
import cv2
import glob
import pandas as pd
import os

with open('time.txt') as f:
    my_times = f.readlines()
    

df = pd.DataFrame(columns=['File_name','Avg_red','Avg_green','Avg_blue','Time','Label'])

#select the path
path = "D:\Python LCC data\L1\*.*"
count =0

for file in sorted(glob.glob(path), key=os.path.getmtime):
    img= cv2.imread(file)
    x,y,c=img.shape
    total_pixel =x*y
    avgr =0
    avgg =0
    avgb =0
    for i in range(0, x):
        for j in range(0, y):
            b,g,r=img[i,j]
            avgr = avgr + r
            avgg = avgg + g
            avgb = avgb + b
    avgr = avgr / total_pixel
    avgg = avgg / total_pixel
    avgb = avgb / total_pixel
    time = my_times[count]
    time = time.replace(':','')
    time = time.replace('\n','')
    count=count+1
    df = df.append({'File_name':file,'Avg_red':avgr,'Avg_green':avgg,'Avg_blue':avgb,'Time':time,'Label':1}, ignore_index=True)
    
```
